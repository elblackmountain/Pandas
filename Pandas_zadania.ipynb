### Pandas - Zadanie 2 - Najczęstsze nazwisko

#### Osób o jakim nazwisku jest najwięcej ?

```python
dataset['nazwisko'].value_counts()[dataset['nazwisko'].value_counts() == dataset['nazwisko'].value_counts().max()]
```
