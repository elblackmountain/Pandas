### Pandas - Zadanie 3 - Płeć

#### Po kolumnie z nazwiskiem dodaj kolumnę reprezentującą płeć danej osoby. Można wykorzystać fakt, że imiona kobiet kończą się literka 'a'.

##### Rozwiązanie 1

```python
def plec1(x):  # w argumencie zawsze zewnętrzna funkcji, moze być tez lambda
    if x[-1]=='a':
        return 'K'
    else:
        return 'M'
```

```python
dataset['płeć'] = dataset['imię'].apply(plec1) # w apply w argumencie zawsze zewnętrzna funkcji, moze być tez lambda
```

```python
#potem trzeba zmienić kolejnośc kolumn
dataset = dataset[['imię', 'nazwisko','płeć', 'rok urodzenia', 'miesiac urodzenia', 'dzień urodzenia','miejsce urodzenia', 'miasto, w którym znajduje się oddział firmy', 'departament, w którym pracuje dana osoba', 'rok roczęcia pracy w firmie','miesiac rozpoczęcia pracy','dzień rozpoczęcia pracy','doswiadczenie (w latach)']]
```

```python
dataset.head(3)
```

##### Rozwiązanie 2

```python
dataset.drop('płeć', axis=1, inplace = True)
```

```python
def plec2(kolumna):
    lista_plci = []
    for i in range(len(kolumna)):
        if kolumna[i][-1] == 'a':
            lista_plci.append('K')
        else:
            lista_plci.append('M')
    
    return lista_plci
```

```python
dataset.insert(2, 'płeć', plec2(dataset.imię))
```

```python
dataset.head(3)
```

##### Rozwiązanie 3

```python
plec3 =  dataset['imię'].apply(lambda imię: imię[-1])=='a'
dataset.insert(2, 'płeć', plec3)
dataset['płeć'] = dataset['płeć'].replace(True, "K")
dataset['płeć'] = dataset['płeć'].replace(False, "M")
```

###  Zadanie 5 - Doświadczenie

##### Do zbioru danych należy dodać kolumnę, która będzie mówiła o poziomie doświadczenia danej osoby, w sposób słowny, przyjmując:
* Junior (do 3 lat doświadczenia) 
* Mid (4 do 8 lat) 
* Senior (9 do 12 lat) 
* Ekspert (od 13 lat)

```python
def define_experience(kolumna):
    list_of_exp = []
    for x in range(len(kolumna)):
        if kolumna[x] <=3:
            list_of_exp.append("Junior")
        elif 4<= kolumna[x] <=8:
            list_of_exp.append("Mid")
        elif 9 <= kolumna[x] <= 12:
            list_of_exp.append("Senior")
        else:
            list_of_exp.append("Ekspert")
    return list_of_exp
```

```python
dataset.insert(13,'poziom', define_experience(dataset['doswiadczenie (w latach)']))
```

```python
dataset.head(3)
```

```python
dataset.drop('poziom', axis=1, inplace = True)
```
