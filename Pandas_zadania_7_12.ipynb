###  Zadanie 8 - Nazwy departamentów

##### Szef chciałby zmienić nazewnictwo departamentów w całej firmie. Napisz kod, który zamieni wartości w kolumnie reprezentującej departament, nastepująco:

* Software Development -> Software Development Team
* HR -> Human Resources
* Data Science -> Data Analytics

```python
def change_name(x):
    if x == 'Software Development':
        return 'Software Development Team'
    elif x == 'HR':
        return 'Human Resources'
    elif x == "Data Science":
        return "Data Analytics"
    else:
        return x

dataset['departament, w którym pracuje dana osoba'] = dataset['departament, w którym pracuje dana osoba'].apply(change_name)
dataset.head()
```

#### metoda .replace()

```python
dataset.replace({'Software Development': 'Software Development Team', 'HR':'Human Resources', 'Data Science': 'Data Analytics'}).head()
```
