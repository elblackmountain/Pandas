##### Który oddział firmy ma najwięcej Data Scientistów ?

```python
import pandas as pd
dataset = pd.read_csv('dataset_pandas_pd3.csv', index_col = 0)
dataset.head(3)
```

```python
data = dataset[dataset['departament, w którym pracuje dana osoba'] 
               == "Data Science"].groupby(['miasto, w którym znajduje się oddział firmy']).count()['departament, w którym pracuje dana osoba']
data[data.idxmax()],data.idxmax()
```
